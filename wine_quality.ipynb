```python
 X=df.drop('quality',axis=1)
print(X)
```

```python
Y = df['quality'].apply(lambda y_val: 1 if y_val>=7 else 0)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
md = RandomForestClassifier()
```

```python
md.fit(X_train,Y_train)
```

```python
X_test_pred = md.predict(X_test)
test_dt_acc = accuracy_score(X_test_pred, Y_test)
```

```python
print('Accuracy: ',test_dt_acc)
```

```python
in_dt = (7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)
npa = np.asarray(in_dt)
#reshaping
inr =npa.reshape(1,-1)

pred = md.predict(inr)
print(pred)

if(pred==1):
    print('Good quality wine')
else:
    print('Bad quality Wine')
```
